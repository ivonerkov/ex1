#<font color='orange'>Направете претпроцесирање на податоците</font>

```python
import pandas as pd
```

```python
data = pd.read_csv('crypto_data.csv')
```

```python
data
```

```python
data['Close (USDT)']
```

```python
matrix = []
for i in range(len(data['Close (USDT)'])-1, 2, -1):

  day2=data['Close (USDT)'][i-2]
  day1=data['Close (USDT)'][i-1]
  matrix.append([data['Close (USDT)'][i], day1, day2, data['Close (BNB)'][i], data['Close (BTC)'][i], data['Close (ETH)'][i]])

matrix
```

```python
df = pd.DataFrame(matrix,columns=['Close','One day ago', 'Two day ago', 'close BNB','close BTC', 'close eth'])
```

```python
df
```

#<font color='orange'>Поделето го на тренирачко и тестирачко множество во сооднос 80:20</font>

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(df[df.columns[1:]],  df['Close'], test_size=0.2)
```

#<font color='orange'>Искористете соодветен Boosting модел за проблемот</font>

```python
#јас ке го искористам XGBOOST моделот за регресија
```

```python
from xgboost import XGBRegressor

model = XGBRegressor()
model.fit(X_train, Y_train)

y_pred = model.predict(X_test)
```

#<font color='orange'>Тестирајте го моделот на тестирачкото множество со метриките за евалуација на дадениот проблем</font>

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
print(mean_squared_error(y_pred, Y_test))
```

```python
r2_score(Y_test, y_pred)
```
