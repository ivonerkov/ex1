#<font color='white'>Користење на соодветен трансформер модел за проблемот</font>

```python
! pip install transformers
```

```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification
```

```python
tokenizer = AutoTokenizer.from_pretrained('bert-base-uncased')
```

```python
model = AutoModelForSequenceClassification.from_pretrained('bert-base-uncased')
```

```python
!pip install datasets
```

```python
from datasets import load_dataset
```

```python
dataset = load_dataset("csv", data_files={"train": "train.csv", "test": "test.csv"})
```

```python
dataset
```

```python
encoded_dataset = dataset.map(lambda t: tokenizer(t['review_es'],  truncation=True), batched=True,load_from_cache_file=False)
```

```python
model = AutoModelForSequenceClassification.from_pretrained('bert-base-uncased', num_labels =2)
```

```python
encoded_dataset
```

#<font color='white'>Фино го подесуваме(тренираме) моделот </font>

```python
from transformers import TrainingArguments, Trainer
```

```python
arg = TrainingArguments(
    "sentiment",
    learning_rate=5e-5,
    num_train_epochs=30,
    per_device_eval_batch_size=10,
    per_device_train_batch_size=10,
    seed=19
)
```

```python
trainer = Trainer(
    model=model,
    args=arg,
    tokenizer=tokenizer,
    train_dataset=encoded_dataset['train'],
    eval_dataset=encoded_dataset['test']
)
```

```python
trainer.train()
```

#<font color='white'>Тестирање на моделот на тестирачкото множество</font>

```python
y_pred = trainer.predict(encoded_dataset['test'])
```

```python
y_pred = y_pred.predictions
```

```python
import numpy as np
```

```python
y_pred = [np.argmax(y_pred[i]) for i in range(0,len(y_pred))]
```

#<font color='white'>Пресметка на f1 score, precission, recall и confusion matrix</font>

```python
from sklearn.metrics import f1_score, classification_report, confusion_matrix
```

```python
f1_score(df_test['label'],y_pred)
```

```python
print(classification_report(df_test['label'],y_pred))
```

```python
print(confusion_matrix(df_test['label'],y_pred))
```
